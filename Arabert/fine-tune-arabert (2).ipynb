```python
import random

# Positive examples (relevant pairs)
positive_pairs = df[['title', 'content']].copy()
positive_pairs['label'] = 1

# Negative examples (irrelevant pairs)
def generate_negative_pairs(df):
    negative_pairs = []
    for idx, row in df.iterrows():
        random_idx = random.choice(df.index)
        if random_idx != idx:
            negative_pairs.append({'title': row['title'], 'content': df.loc[random_idx, 'content'], 'label': 0})
    return pd.DataFrame(negative_pairs)

negative_pairs = generate_negative_pairs(df)

# Combine positive and negative pairs
all_pairs = pd.concat([positive_pairs, negative_pairs]).reset_index(drop=True)

# Shuffle the dataset
all_pairs = all_pairs.sample(frac=1).reset_index(drop=True)
```

```python
all_pairs.rename(columns={"title": "query", "content": "document"}, inplace=True)
```

```python
from sklearn.model_selection import train_test_split
train_data, val_data = train_test_split(all_pairs, test_size=0.2, random_state=42)
```

```python
from transformers import AutoTokenizer

# Load AraBERT tokenizer
tokenizer = AutoTokenizer.from_pretrained("aubmindlab/bert-base-arabertv2")

# Tokenize the data
def tokenize_function(example):
    return tokenizer(
        example['query'], 
        example['document'], 
        truncation=True, 
        padding="max_length", 
        max_length=512
    )
```

```python
from datasets import Dataset

# Convert Pandas DataFrame to Dataset
train_dataset = Dataset.from_pandas(train_data)
val_dataset = Dataset.from_pandas(val_data)

# Tokenize datasets
train_dataset = train_dataset.map(tokenize_function, batched=True)
val_dataset = val_dataset.map(tokenize_function, batched=True)
```

```python
from transformers import AutoModelForSequenceClassification

# Load AraBERT
model = AutoModelForSequenceClassification.from_pretrained("aubmindlab/bert-base-arabertv2", num_labels=2)
```

```python
import evaluate

accuracy = evaluate.load("accuracy")

def compute_metrics(eval_pred):
    logits, labels = eval_pred
    predictions = logits.argmax(axis=-1)
    return accuracy.compute(predictions=predictions, references=labels)
```

```python
from transformers import TrainingArguments

training_args = TrainingArguments(
    output_dir="./arabert_ir_finetuned",  # Save directory
    eval_strategy="epoch", 
    save_strategy="epoch",  
    logging_strategy="epoch",
    learning_rate=5e-5,
    per_device_train_batch_size=16,
    per_device_eval_batch_size=16,
    num_train_epochs=3,
    weight_decay=0.01,
    logging_dir="./logs",
    save_total_limit=2,
    load_best_model_at_end=True,
    disable_tqdm=False,
    report_to="none"
)
```

```python
from transformers import Trainer

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=val_dataset,
    tokenizer=tokenizer,
    compute_metrics=compute_metrics,
)
```

```python
trainer.train()
```

```python
trainer.save_model("arabert_ir_finetuned")
tokenizer.save_pretrained("arabert_ir_finetuned")
```

```python
trainer.evaluate()
```
